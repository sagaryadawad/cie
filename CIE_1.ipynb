```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('/content/mtcars.csv')
print(df.head())
plt.hist(x='mpg',data=df,edgecolor='k')
plt.xlabel('mpg')
plt.ylabel('frequency')
plt.title('histogram')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('/content/mtcars.csv')
print(df.head())
plt.bar(x='carb', height='mpg', data=df)
plt.xlabel('carb')
plt.ylabel('mpg')
plt.title('bar plot')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('/content/mtcars.csv')
print(df.head())
plt.scatter(x='wt', y='mpg', data=df)
plt.xlabel('wt')
plt.ylabel('mpg')
plt.title('scatter plot')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('/content/mtcars.csv')
print(df.head())
plt.boxplot(x='mpg', data=df)
plt.xlabel('mpg')
plt.title('box plot')
plt.show()
```
